```
# Creating widgets for various parameters
dbutils.widgets.text("populationSize", "1000")
dbutils.widgets.text("exporterCSV", "true")
dbutils.widgets.text("exporterFHIR", "false")
dbutils.widgets.text("gender", "")
dbutils.widgets.text("minAge", "0")
dbutils.widgets.text("maxAge", "100")
dbutils.widgets.text("seed", "")
dbutils.widgets.text("referenceDate", "")
dbutils.widgets.text("clinicianSeed", "")
dbutils.widgets.text("state", "")
dbutils.widgets.text("city", "")

# Retrieving parameter values
populationSize = dbutils.widgets.get("populationSize")
exporterCSV = dbutils.widgets.get("exporterCSV")
exporterFHIR = dbutils.widgets.get("exporterFHIR")
gender = dbutils.widgets.get("gender")
minAge = dbutils.widgets.get("minAge")
maxAge = dbutils.widgets.get("maxAge")
seed = dbutils.widgets.get("seed")
referenceDate = dbutils.widgets.get("referenceDate")
clinicianSeed = dbutils.widgets.get("clinicianSeed")
state = dbutils.widgets.get("state")
city = dbutils.widgets.get("city")

# Constructing the command
command = f"java -jar synthea-with-dependencies.jar -p {populationSize} --exporter.csv.export={exporterCSV} --exporter.fhir.export={exporterFHIR}"
if gender:
    command += f" -g {gender}"
if minAge and maxAge:
    command += f" -a {minAge}-{maxAge}"
if seed:
    command += f" -s {seed}"
if referenceDate:
    command += f" -r {referenceDate}"
if clinicianSeed:
    command += f" -cs {clinicianSeed}"
if state:
    command += f" {state}"
if city:
    command += f" {city}"

import os
os.system(command)
```
